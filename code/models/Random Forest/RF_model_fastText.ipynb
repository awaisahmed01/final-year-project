<h3>Final Classifier</h3>

```python
rf_classifier.fit(X_train_ft, y_train)
```

```python
# Prediciting document labels for test data
y_rf_pred_ft = rf_classifier.predict(X_test_ft)

rf_report = classification_report(y_test, y_rf_pred_ft, target_names=multi_names)
print("Report: \n", rf_report)
```

```python
# Generate the confusion matrix


cm = confusion_matrix(y_test, y_rf_pred_ft)

# Display the confusion matrix using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=multi_names, yticklabels=multi_names)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Class')
plt.ylabel('True Class')
plt.show()
```
